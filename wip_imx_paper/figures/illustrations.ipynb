```python
from deepface import DeepFace
import matplotlib.pyplot as plt
import PIL
import numpy as np

!dir finalphotoface
```

```python
pictures = ["finalphotoface/anger.png", "finalphotoface/disgust.png", "finalphotoface/fear.png", "finalphotoface/happy.png",
            "finalphotoface/sad.png", "finalphotoface/surprise.png", "finalphotoface/neutral.png", "finalphotoface/contempt.png"]

for picture in pictures:
    img = np.array(PIL.Image.open(picture))
    plt.title(picture.split("/")[1].split(".")[0])
    plt.imshow(img)
    plt.show()

    # Predict emotions
    try:
        obj = DeepFace.analyze(img_path=img, actions=(
            'emotion'), detector_backend='mtcnn')
        print(obj)
        print(obj[0]["dominant_emotion"])
    except Exception:
        print("Crashed")
```
